```python
import torch
```

```python
# N matrisini olasılıkları kaydetmek için kullanıcaz
# ingilizce alfabesinde 26 harf bulunmaktadır ve biz başlangıç ve sonu ifade etmesi için '.' kullanıcaz
# yani boyutumuz 27x27 olmalı

N = torch.zeros((27, 27), dtype=torch.int32) 
```

```python
# kelimeleri yan yana yazdırıp set ile aynı harfleri tekrar eklemiyoruz sort ile 
# alfabetik sıraya uygun düzenliyoruz

# stoi listesi harflerin indeksini
# itos listesi indeksteki harfleri tutar

chars = sorted(list(set(''.join(words)))) 
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s,i in stoi.items()}
```

```python
# sıklıklar hesaplanıyor
for w in words:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        N[ix1, ix2] += 1
```

```python
# bu grafik bize harflerin ardı ardına kaç tane geldiğini gösterir
# örneğin .(başlangıç) dan sonra a harfininin 4410 kere geldiğini açıkça gösterir 

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(16,16))
plt.imshow(N, cmap='Blues')
for i in range(27): 
    for j in range(27): 
        chstr = itos[i] + itos[j]
        plt.text(j, i, chstr, ha="center", va="bottom", color="gray") 
        plt.text(j, i, N[i, j].item(), ha="center", va="top", color="gray")
plt.axis=('off')
```

```python
# P matrisini her satırın kendi satırın toplamına bölünerek olasılıklar çıkarılır
# Önemli nokta P nin tüm N matrisi toplamına değil de ilgili satırın toplamına bölünmesidir

P = (N+1).float() # N+1 işlemi tüm değerlere 1 ekler bu bir yumuşatma işlemidir ilerde anlaşılacaktır
P = P / P.sum(1, keepdim=True)
```

```python
# Biagram ile yeni isimler üretmeye çalıştığımızda sonuçlar görüldüğü gibidir

g = torch.Generator().manual_seed(2147483647)

for i in range(5): # 5 adet örnek isim üretsin
    out = []
    ix = 0
    while True:
        
        p = P[ix]
        ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
        out.append(itos[ix])
        if ix == 0:
            break
    print(''.join(out))
```

```python
# bu bizim loss hesapladığımız kısımdır
# olasılıklara log(x) uygularız bu bize bir kelime olasılığını çarpmak yerine toplama imkanı verir
# log(x) bize negatif değer döndürecek 0 ve 1 aralığında 
# 0 a yakınlığı büyüklüğünü etkilicek örneğin log(1) = 0 log(0.1) ≈ -2.3
# pozitife çevirdiğimizde düşük olasıklı ihtimalın loss değeri daha büyük olacaktır buda modelimize ceza olarak düşünebiliriz
# amacımız olasılığı arttırarak loss değerini en az tutmaktır

log_likehood = 0.0
n = 0
for w in words:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        prob = P[ix1, ix2]
        logprob = torch.log(prob)
        log_likehood += logprob
        n += 1
        #print(f"{ch1}{ch2}: {prob:.4f} {logprob:.4f}")

print(f"{log_likehood=}")
nll = -log_likehood
print(f"{nll=}")
print(f"{nll/n}") # bu tüm listemizdeki ortalama loss değeridir
```

```python
# Bigram ile tahminde bulunduk ve sonuçları gördük ama biz model eğitip tahminde bulunmak istiyoruz
# bunun içinde veri seti hazırlamalıyız

xs, ys = [], []

for w in words[:1]:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        print(ch1, ch2)
        xs.append(ix1)
        ys.append(ix2)

xs = torch.tensor(xs)
ys = torch.tensor(ys)
print("xs:", xs)
print("ys:", ys)
```

```python
# xs dizisine one hot encoder uyguluyoruz 

import torch.nn.functional as F
xenc = F.one_hot(xs, num_classes=27).float()
print("xenc:", xenc)
print("xenc boyutu:", xenc.shape)
```

```python
plt.imshow(xenc)
```

```python
# burada ağrılık matrisini oluşturarak modelimizin tahminde bulunuyor
# ve loss değerini yazdırıyoruz

W = torch.randn((27,27), generator=g, requires_grad=True)
logits = xenc @ W # log-counts
counts = logits.exp() # equivalent N
print(counts.shape)
probs = counts / counts.sum(1, keepdims=True)
```

```python
# bigram ile loss hesaplamasını detaylıca görebilirsiniz

nlls = torch.zeros(5)
for i in range(5):
    x = xs[i].item()
    y = ys[i].item()
    print('----------------------')
    print(f"biagram example {i+1}: {itos[x]}{itos[y]} (indexes {x},{y}")
    print("input to the neural net:", x)
    print("output probabilities from the neural net:", probs[i])
    print("label (actual next character):",y)
    p = probs[i, y]
    print("probabilities assigned by the net to the correct character:", p.item())
    logp = torch.log(p)
    print("log likehood:", logp.item())
    nll = -logp
    print("negative log likehood:", nll.item())
    nlls[i] = nll

    print("=================")
    print("average negative log likehood, i.e. loss =", nlls.mean().item())
```

```python
# veri seti oluşturalım

xs, ys = [], []

for w in words:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        xs.append(ix1)
        ys.append(ix2)

xs = torch.tensor(xs)
ys = torch.tensor(ys)
num = xs.nelement()

# ağırlık matrisini oluşturarak sinir ağımızı başlatalım
W = torch.randn((27,27), generator=g, requires_grad=True)
```

```python
# tüm süreci toparlayalım
# bu kısımda ileri yayılım uygulanmaktadır

import torch.nn.functional as F
for i in range(50):
    xenc = F.one_hot(xs, num_classes=27).float()
    logits = xenc @ W
    counts = logits.exp()
    probs = counts / counts.sum(1, keepdims=True)
    loss = -probs[torch.arange(num),ys].log().mean() + 0.01*(W**2).mean()
    print(loss.item())

    
    # geri yayılım gerçekleştiriliyor

    W.grad = None # her geri yayılımdan önce grad değerlerinin sıfırlanması önemlidir
    loss.backward()

    
    # W değerleri güncellenir
    
    W.data += -50 * W.grad # öğrenme oranını uygun şekilde ayarlarız bu örnek basit bir sinir ağı olduğu için
                            # öğrenme oranını epeyce yüksek vericez
```

```python
# modelimizi kullanarak kelime türetelim

g = torch.Generator().manual_seed(2147483647)

for i in range(5):
    out = []
    ix = 0
    while True:
        
        xenc = F.one_hot(torch.tensor([ix]), num_classes=27).float()
        logits = xenc @ W
        counts = logits.exp()
        p = counts / counts.sum(1, keepdims=True)

        ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
        out.append(itos[ix])
        if ix == 0:
            break
    print(''.join(out))
```
